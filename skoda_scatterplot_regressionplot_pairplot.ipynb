```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
```

```python
df = pd.read_csv("F:\\python\Data_Visualization_with_Python_s2-main\HW-2\\skoda.csv")
display(df.head(5))
```

# Scatterplot with numerical column:

1)mileage vs price 2)mpg vs price 3)engineSize vs price


```python
plt.figure(figsize=(20,15))

plt.subplot(2, 2, 1)
sns.scatterplot(x="mileage", y="price", data=df,hue="fuelType")

plt.subplot(2, 2, 2)
sns.scatterplot(x="mpg", y="price", data=df,hue="fuelType")

plt.subplot(2, 2, 3)
sns.scatterplot(x="engineSize", y="price", data=df,hue="fuelType")

plt.tight_layout()
plt.show()
```

# Regression plot with numerical columns:

* mileage vs price
* mpg vs price
* engineSize vs price

```python
plt.figure(figsize=(20,15))

plt.subplot(2, 2, 1)
sns.regplot(x="mileage", y="price", data=df, line_kws={"color":"red"})

plt.subplot(2, 2, 2)
sns.regplot(x="mpg", y="price",data=df,line_kws={"color":"blue"})

plt.subplot(2, 2, 3)
sns.regplot(x="engineSize", y="price", data=df,line_kws={"color":"green"})

plt.tight_layout()
plt.show()
```

# Pairplot with numerical columns:

Non-categorical


```python
sns.pairplot(df, corner=True)
plt.show()
```

Categorical:
